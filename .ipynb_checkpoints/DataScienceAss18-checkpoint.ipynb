```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
import graphviz
import numpy as np
```

```python
data = pd.read_csv('indian_food.csv', sep=',')
data = data.dropna()
data.head()
```

```python
data_train, data_test = train_test_split(data, test_size=0.3, random_state=42, stratify=data['region'])
print(data_train.shape, data_test.shape)
```

```python
data.corr()
```

```python
features= ['prep_time'] # add 'petal_width' ('species' does not work; categorical is not implemented in DT of sciki learn)
dt_regression = DecisionTreeRegressor(max_depth = 5) # Increase max_depth to see effect in the plot
dt_regression.fit(data_train[features], data_train['cook_time'])
```

```python
trainprediction = dt_regression.predict(data_train[features])
testprediction = dt_regression.predict(data_test[features])
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)
```

```python
predictionsOnTrainset = dt_regression.predict(data_train[features])
predictionsOnTestset = dt_regression.predict(data_test[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, data_train['cook_time'])
rmseTest = calculate_rmse(predictionsOnTestset, data_test['cook_time'])

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```

```python
def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
plot_tree_regression(dt_regression, features)
```
